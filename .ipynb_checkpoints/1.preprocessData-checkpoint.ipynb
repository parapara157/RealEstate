```python
#2018년 국교부서울특별시아파트실거래가 데이터
data2018=pd.read_csv('data/2018.csv',encoding='euc-kr')                                             
```

```python
data2018.columns=data2018.iloc[15]
data2018=data2018[16:]
data2018=data2018.reset_index()
del data2018['index']
del data2018.columns.name
```

```python
#공백없애기,콤마지우기,숫자변환이 필요함
data2018['거래금액(만원)'][0]
```

```python
def removeComma(data):
    return "".join(data.split()[0].split(','))
```

```python
data2018['거래금액(만원)']=data2018['거래금액(만원)'].apply(removeComma).astype('float')
```

```python
#구,동,월별로 따로 변수를 만들어 줌.
def changeSomeData(data):
    data['big']=data['시군구'].apply(lambda e:e.split(' ')[1])
    data['small']=data['시군구'].apply(lambda e:e.split(' ')[2])
    del data['계약일']
    def makingMonth(date):
        date=str(date)
        return str(date[4])+str(date[5])
    data['month']=data['계약년월'].apply(makingMonth)
    return data
```

```python
data2018=changeSomeData(data2018)
```

```python
#위치특성을 이용하기 위해 카카오맵 API를 이용하여 위도와 경도를 넣어주겠다.
data2018['latitude']=1.1
data2018['longitude']=1.1
```

```python
#사전에 찾은 잘못된 주소들
data2018.loc[data2018['단지명']=='천왕이펜하우스1단지','도로명']='천왕로 91'    
data2018.loc[data2018['단지명']=='래미안포레','본번']='361'
data2018.loc[data2018['단지명']=='강남한양수자인(4단지)','본번']='77'
data2018.loc[data2018['단지명']=='강남한신휴플러스8단지','본번']='217'
data2018.loc[data2018['단지명']=='강남한신휴플러스6단지','본번']='336'
data2018.loc[data2018['단지명']=='천왕이펜하우스2단지','도로명']='천왕로 92'
data2018.loc[data2018['단지명']=='천왕이펜하우스3단지','도로명']='천왕로 56'
data2018.loc[data2018['단지명']=='천왕이펜하우스4단지','도로명']='천왕로 29'
data2018.loc[data2018['단지명']=='천왕이펜하우스5단지','도로명']='천왕로 10'
data2018.loc[data2018['단지명']=='천왕이펜하우스6단지','도로명']='천왕로 9'
data2018.loc[data2018['단지명']=='코지한주','도로명']='덕릉로123길 53'
```

```python
def getPlace(start,train):
    
    def getLatLng(addr):                                    ###give a address and then get the latitude, longitude
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+addr
        headers = {"Authorization": "API신청후 얻은 KEY를 입력"}
        result = json.loads(str(requests.get(url,headers=headers).text))
        match_first = result['documents'][0]['address']
        return float(match_first['y']),float(match_first['x'])

    def getLatLng2(addr):                                   ###도로명으로 반환할경우 값이  반환형태가 아주 약간다름
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+addr
        headers = {"Authorization": "API신청후 얻은 KEY를 입력"}
        result = json.loads(str(requests.get(url,headers=headers).text))
        match_first = result['documents'][0]                ###요차이 밖에 없다
        return float(match_first['y']),float(match_first['x'])

    def getLatLng3(addr):    ###오류 확인활때 전체 값 볼려고 
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+addr
        headers = {"Authorization": "API신청후 얻은 KEY를 입력"}
        result = json.loads(str(requests.get(url,headers=headers).text))
        return result

    for i in range(start,len(train)):                       ###에러로 인해서 다시 해야될때 range옆에 숫자로 시작지점 조절
        try:
            place=getLatLng(str(train['시군구'][i])+' '+str(train['본번'][i])+'-'+str(train['부번'][i]))
            train['latitude'][i]=place[0]
            train['longitude'][i]=place[1]
        except IndexError :
            try:
                place=getLatLng(str(train['시군구'][i])+' '+str(train['번지'][i]))
                train['latitude'][i]=place[0]
                train['longitude'][i]=place[1]
            except IndexError :
                try:
                    place=getLatLng2(str(train['시군구'][i])+' '+str(train['도로명'][i]))
                    train['latitude'][i]=place[0]
                    train['longitude'][i]=place[1]
                except IndexError:
                    pass
    return train
```

```python
#이런식으로 위도와 경도를 넣어줄 수 있다.
#다만 하루에 얻을 수 있는 정보의 양이 제한되어 있기 때문에 함수의 start값을 조절하여
#위도와 경도를 계속 넣어줄 수 있도록 한다.
getPlace(0,data2018)
```
