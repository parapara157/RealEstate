```python
import pandas as pd
import json
import requests
import copy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data2018=pd.read_csv('dataAfterPreprocess/data2018') 
data2017=pd.read_csv('dataAfterPreprocess/data2017')
data2016=pd.read_csv('dataAfterPreprocess/data2016') 
data2015=pd.read_csv('dataAfterPreprocess/data2015') 
data2014=pd.read_csv('dataAfterPreprocess/data2014') 
data2013=pd.read_csv('dataAfterPreprocess/data2013') 
data2012=pd.read_csv('dataAfterPreprocess/data2012') 
data2011=pd.read_csv('dataAfterPreprocess/data2011') 
data2010=pd.read_csv('dataAfterPreprocess/data2010') 
```

```python
train=pd.concat([data2018,data2017,data2016,data2015,data2014,data2013,data2012,data2011,data2010],axis=0)
train=train.reset_index()
del train['index']
train.shape
```

```python
###경제상황을 반영해주기 위해 각종 지수들을 넣어줌.

###공동주택 실거래 가격지수 아파트 
apartIndex=pd.read_csv("indexData/apartIndex.csv",encoding="cp949")
apartIndex=list((apartIndex.iloc[10].reset_index())[10])[2:]
###경기종합 선행 지수
economicIndex=pd.read_csv("indexData/economicPredictionIndex.csv",encoding="cp949")
economicIndex=list((economicIndex.iloc[10].reset_index()[10])[1:])
###생산자 물가 지수
produceIndex=pd.read_csv("indexData/produceIndex.csv",encoding="cp949")
produceIndex=list((produceIndex.iloc[10].reset_index()[10])[1:])
###소비자 물가 지수
spendIndex=pd.read_csv("indexData/spendIndex.csv",encoding="cp949")
spendIndex=list((spendIndex.iloc[10].reset_index()[10])[1:])
```

```python
del train['시군구']
del train['번지']
del train['본번']
del train['부번']
del train['단지명']   
del train['도로명']   
del train['small']
###object에서 category로 바꾸어주어야함
train['big']=train['big'].astype('category')
```

```python
###2018/12데이터를 예측하므로 2018/11까지 있어야됨
###2018/12데이터에는 2018/11데이터넣음
###2010/01데이터에는 2009/12데이터넣음
def makeApartIndex(x):
    index=(x-201001)//100*12+(x-201001)%100
    return apartIndex[index]
def makeEconomicIndex(x):
    index=(x-201001)//100*12+(x-201001)%100
    return economicIndex[index]
def makeSpendIndex(x):
    index=(x-201001)//100*12+(x-201001)%100
    return spendIndex[index]
def makeProduceIndex(x):
    index=(x-201001)//100*12+(x-201001)%100
    return produceIndex[index]
```

```python
train['apartIndex']=train['계약년월'].apply(makeApartIndex).astype('float')
train['economicIndex']=train['계약년월'].apply(makeEconomicIndex).astype('float')
train['spendIndex']=train['계약년월'].apply(makeSpendIndex).astype('float')
train['produceIndex']=train['계약년월'].apply(makeProduceIndex).astype('float')
```

```python
train.columns
```

```python
train.columns=['USM','Sold date','Sold price','Storey','Build year','latitude','longitude','District','Sold month','Apart index'
               ,'Leading index','Consumer index','Producer index']
```

```python
#예측할시에 단위면적당 가격으로 예측해줌.
train['PPSM']=train['Sold price']/train['USM']
```

```python
###상관관계를 알아보자
###다중 공성선의 존재 확인 가능
corr=train.corr()
corr.style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
grouped=train['PPSM'].groupby(train['District'])
big=pd.DataFrame(grouped.mean())
```
